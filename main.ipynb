```python
import pandas as pd
import numpy as np

# Loading the data required
dt = pd.read_csv('./data/answer_key.csv', dtype=str)
# data

# This is unorder dict
data1: dict[str, list[str, str]] = {}
# This is ordered dict
data: dict[str, list[str, str]] = {}
dt = dt.to_dict()

keys = list(dt.keys())
i = 0
while i < len(dt[keys[0]]):
    data1[dt[keys[0]][i]] = [dt[keys[1]][i], None]
    i += 1

# Now we have the main data as a key value pair
# Question ID : Option ID / Numerical Answer, Student Answer
```

```python
from pyquery import PyQuery as pq
h_data = pq(filename='./data/question_paper_html')

tables = h_data('td.rw table.menu-tbl')
q_tables = h_data('td.rw table.questionRowTbl')

i = 0
for table in tables:
    tbl = pq(table)
    typ = tbl('tr').eq(0)("td").eq(1).text()
    # Question ID
    q_id = tbl('tr').eq(1)("td").eq(1).text()
    # Answer Status
    if typ == "MCQ":
        status = tbl('tr').eq(6)("td").eq(1).text()
    else:
        status = tbl('tr').eq(2)("td").eq(1).text()
    # Option Number
    awns_number = 0
    # Option ID
    awns_id = None
    if status == "Answered":
        if typ == "MCQ":
            awns_number = int(tbl('td').eq(15).text())
            awns_id = tbl('tr').eq(awns_number+1)("td").eq(1).text()
        else:
            awns_id = q_tables.eq(i).eq(0).text().split("\n")[-1]
    data[q_id] = [data1[q_id][0], awns_id]
    # print([typ, q_id, status])
    i += 1
data
```

```python
# Calculating the score

score = 0

i = 0
correct = 0
incorrect = 0
for x in data:
    dt = data[x]
    i += 1
    if not dt[1]:
        # Unanswered question are not evaluated
        continue
    if dt[0] == dt[1]:
        # print(f"Question {i+1} is correct")
        correct += 1
        score += 4
    else:
        # print(f"Question {i+1} is incorrect")
        incorrect += 1
        score -= 1
# score
print("Unanswered Questions: ", i-correct-incorrect)
print("Attempted Questions: ", correct+incorrect)
print("Correct Questions: ", correct)
print("Incorrect Questions: ", incorrect)
print("Score: ", score)
```
